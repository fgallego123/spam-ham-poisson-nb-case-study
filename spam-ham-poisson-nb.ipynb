```python
#ttest of two diff samples
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, uniform, norm 
from scipy.stats import norm, t, ttest_1samp, ttest_ind
```

```python
df = pd.read_csv('spam.csv', sep = ",",  encoding = 'latin-1')
df.head()
outcomes = df['v1']
```

```python
from scipy.stats import poisson
```

```python
#binarize outcomes
bin_outcomes = []
for outcome in outcomes:
    if outcome == 'ham':
        bin_outcomes.append(0)
    else:
        bin_outcomes.append(1)

outcomes = bin_outcomes
outcomes
```

```python
#Hypothesis testing 
mu0 = .125
alpha = 0.025
n = len(outcomes)
degfree = len(outcomes) - 1
sum(outcomes)
```

```python
x_bar = sum(outcomes)/len(outcomes) #sample mean is .134
```

```python
std = np.std(outcomes, ddof =1)
```

```python
#use z test, always use z test for proportions

```

```python
prop = np.sqrt((mu0)*(1-mu0)/n)
```

```python
sum(outcomes)
```

```python
rv = norm(mu0, prop)
```

```python
1 - rv.cdf(x_bar) #reject since right tailed and less than alpha
```

```python
#Poisson
l = sum(outcomes)/28
```

```python
rvp = poisson(l) 
```

```python
1 - rvp.cdf(29) #prob of at least 30 spam per day
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
count_vect = CountVectorizer()
tfidf_transformer = TfidfTransformer()

counts = count_vect.fit_transform(df['v2'])  # change
tfidfs = tfidf_transformer.fit_transform(counts)

X_train, X_test, y_train, y_test = train_test_split(tfidfs, df['v1'], test_size=0.33, random_state=42) #change

nb = MultinomialNB().fit(X_train, y_train)
predictions = nb.predict(X_test)
```

```python
len(predictions)
```

```python
len(y_test)
```

```python
raw = {'actual' : y_test, 'pred' : predictions}
accuracy_df = pd.DataFrame(data = raw)
```
